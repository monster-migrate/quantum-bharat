```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
df = pd.read_csv("./datasets/cleaned.csv")
```

```python
df.info()
```

```python
ax = sns.regplot(data=df, x='CONDUCTIVITY (µmhos/cm)-Mean', y="pH-Mean")
ax.set_xscale('log')
```

```python
scaler = MinMaxScaler()
df[['cond_norm', 'ph_norm']] = scaler.fit_transform(df[['CONDUCTIVITY (µmhos/cm)-Mean', 'pH-Mean']])

sns.regplot(data=df, x='cond_norm', y='ph_norm')
plt.show()
```
